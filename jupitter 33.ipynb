```python
import pyautogui as gui
import time
n=int(gui.prompt("Enter the no of files you want to rename"))
ext=gui.prompt('Enter the extension you want to add')
gui.confirm("The Bot is going to start in 5 sec...")
time.sleep(5)
gui.click()
while(n>0):
    gui.hotkey('fn','f2')
    gui.press('right',interval=0.0000001)
    gui.write(ext,interval=0.0000001)
    gui.press('enter',interval=0.0000001)
    gui.press('down',interval=0.0000001)
    n-=1
```
